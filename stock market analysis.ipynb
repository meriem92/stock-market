```python
import pandas as pd
```

```python
apple = pd.read_csv('AAPL.csv')
facebook = pd.read_csv('FB.csv')
netflix = pd.read_csv('NFLX.csv')
amazon = pd.read_csv('AMZN.csv')
microsoft = pd.read_csv('MSFT.csv')
paypal = pd.read_csv('PYPL.csv')
```

```python
amazon.head()
```

```python
paypal.head()
```

```python
# Creating a list of dataframes
dfs = [apple, facebook, netflix, amazon, microsoft, paypal]
```

```python
# Creating moving average 50 and Moving average 200
for df in dfs:
    df['MA50'] = df.Close.rolling(50).mean()
    df['MA200'] = df.Close.rolling(200).mean()
```

```python
paypal.head(50)
```

```python
# Creating column previous day close price
for df in dfs:
    df['Previous day close price'] = df.Close.shift(1)
```

```python
paypal.head()
```

```python
# Creating a column change in price  ( if it's negative then price decreased , if it's positive then price increased)
for df in dfs:
    df['Change in price'] = df['Close'] - df['Previous day close price']
```

```python
paypal.head()
```

```python
# Calculate the percentage Change / return   formula is (Current-Previous/Previous) * 100.
for df in dfs:
    df['Percentage change in price'] = df.Close.pct_change()
```

```python
paypal.head()
```

```python
# Creating column previous day volume
for df in dfs:
    df['Previous day volume'] = df.Volume.shift(1)
```

```python
paypal.head()
```

```python
# Creating a column change in volume
for df in dfs:
    df['Change in volume'] = df['Volume'] - df['Previous day volume'] 
```

```python
paypal.head()
```

```python
# Creating Percentage change in volume
for df in dfs:
    df['Percentage change in volume'] = df.Volume.pct_change()
```

```python
paypal.head()
```

```python
# Saving dataframes as csv files to send it to Tableau 
apple.to_csv('Apple.csv')
paypal.to_csv('Paypal.csv')
netflix.to_csv('Netflix.csv')
microsoft.to_csv('Microsoft.csv')
amazon.to_csv('Amazon.csv')
facebook.to_csv('facebook.csv')
```
